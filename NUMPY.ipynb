```python
import numpy as np

# Example array
a = np.array([1, 2.71828183, 7.3890561, 20.08553692, 54.59815003])

# Sum of all elements
print(np.sum(a))    # total sum

# 2D Array
x = np.array([[1, 2],
              [3, 4]])
print(x)

# Column-wise sum
print(x.sum(axis=0))   # [4, 6]

# Row-wise sum
print(x.sum(axis=1))   # [3, 7]

# Minimum and Maximum
print(x.min())   # 1
print(x.max())   # 4

# Index of minimum and maximum
print(x.argmin())  # 0  (position of 1 in flattened array)
print(x.argmax())  # 3  (position of 4 in flattened array)
```
